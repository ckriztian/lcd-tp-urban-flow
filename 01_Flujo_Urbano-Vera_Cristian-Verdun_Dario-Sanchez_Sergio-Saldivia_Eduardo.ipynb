````python

#@title Descargar el dataset raw original y almacenarlo en urban_flow/data/raw.
# URL del dataset
url: str = (
  "https://raw.githubusercontent.com/HAD141/datasets/"
  "refs/heads/main/TrabajosPracticos/urban_flow/speeding_fines.csv"
)

# Ruta destino (estructura ya creada en el repo)
file_path: str = "urban_flow/data/raw/speeding_fines.csv"

# Descarga del dataset
response = requests.get(url)

# Validación de descarga
if response.status_code != 200:
  raise Exception(
    f"Error al descargar el dataset. Código: {response.status_code}"
  )

# Guardado en disco
with open(file_path, "wb") as file:
  file.write(response.content)

# Carga del dataset en DataFrame
df: pd.DataFrame = pd.read_csv(file_path)
````

````python
#@title Visualización de las primeras 5 filas del dataset
df.head()
````
