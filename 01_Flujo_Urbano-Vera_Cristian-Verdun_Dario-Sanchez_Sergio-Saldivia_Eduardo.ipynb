````python
#@title **• Normalización de fechas**

    # Normalización de fechas
print("Datos de fechas normalizadas:")
def normalizar_fechas(df: pd.DataFrame) -> pd.DataFrame:
    """
    Normaliza la columna 'fecha' del dataset:
    - Convierte a datetime
    - Maneja errores
    - Reemplaza valores inválidos
    - Formatea a YYYY-MM-DD
    """

    df_dates: pd.DataFrame = df.copy()

    # Conversión a datetime (formato argentino)
    df_dates["fecha"] = pd.to_datetime(
        df_dates["fecha"],
        errors="coerce",
        dayfirst=True
    )

    # Reemplazo de fechas inválidas
    df_dates["fecha"] = df_dates["fecha"].fillna(
        pd.Timestamp("1932-01-01")
    )

    # Formateo final
    df_dates["fecha"] = df_dates["fecha"].dt.strftime("%Y-%m-%d")

    return df_dates


# Aplicación de la función
df_dates: pd.DataFrame = normalizar_fechas(df)

# Mostrar resultado
df_dates["fecha"].head(10)
````

>Se aplicó un proceso de limpieza y estandarización sobre la columna fecha del dataset con los siguientes pasos:

>Conversión a tipo fecha (datetime) usando pd.to_datetime, considerando formato argentino (dayfirst=True).
Gestión de errores: cualquier valor no interpretable como fecha válida se transforma en NaT (errors="coerce").
Reemplazo de valores inválidos: las fechas nulas (NaT) se sustituyen por la fecha por defecto 1932-01-01.
Formato final: todas las fechas se convierten al estándar internacional YYYY-MM-DD.

````python
#@title **• Normalización de horas al formato 24hs**

# Ver los distintos formatos de hora que hay
print("Datos de horas actuales:")
print(df["hora"].unique()[:20])
````

> Hay tres tipos de formatos:
>
> * **AM/PM: 09:01 AM** → hay que convertirlo a 24hs
> * **Ya en 24hs:** 18:01, 14:01, 08:01 → estos están bien
> * **Inválidos:** invalid_hour → estos van como 00:00

````python
def normalizar_horas(df: pd.DataFrame) -> pd.DataFrame:
  """
  Normaliza la columna 'hora' del dataset:
  - Convierte formatos AM/PM a 24hs
  - Mantiene horas ya en formato 24hs
  - Reemplaza valores inválidos por 00:00
  """
  df_horas: pd.DataFrame = df.copy()

  # Conversión a datetime para unificar formatos
  horas_parseadas: pd.Series = pd.to_datetime(
    df_horas["hora"],
    format="mixed",
    errors="coerce"
  )

  # Formateo a 24hs, inválidas quedan como 00:00
  df_horas["hora"] = horas_parseadas.dt.strftime("%H:%M").fillna(
    "00:00"
  )

  return df_horas


# Aplicación de la función sobre el df ya con fechas normalizadas
df_horas: pd.DataFrame = normalizar_horas(df_dates)

# Mostrar las 10 primeras horas
df_horas["hora"].head(10)
````

>Se normalizó la columna `hora` para unificar su formato y corregir valores inválidos. Para ello, se convirtieron las horas a un formato estándar de 24 horas (HH:MM), se identificaron los valores incorrectos y se reemplazaron por `00:00`, logrando así una columna homogénea y lista para el análisis.

````python
#@title **• Normalización de ubicaciones**

# Ver las ubicaciones únicas
print("Datos sobre ubicaciones actuales:")
print(df_horas["ubicacion"].unique())
print("-------------------------------------------------------------")
````

>La misma ubicación aparece escrita de formas distintas (por ejemplo Av. Siempre Viva, Av Siempre Viva, o AV LIBERTADOR y av libertador). Hay que unificar quitando puntos y pasando todo a mayúsculas.


````python
print("Ubicaciones normalizadas:")
# Clase para normalizar ubicaciones
def normalizar_ubicaciones(df: pd.DataFrame) -> pd.DataFrame:
  """
  Normaliza la columna 'ubicacion' del dataset:
  - Quita caracteres especiales
  - Convierte todo a mayúsculas
  """
  df_ubic: pd.DataFrame = df.copy()

  # Quitar caracteres especiales (deja solo letras, números y espacios)
  df_ubic["ubicacion"] = df_ubic["ubicacion"].str.replace(
    r"[^a-zA-Z0-9\s]", "", regex=True
  )

  # Pasar a mayúsculas
  df_ubic["ubicacion"] = df_ubic["ubicacion"].str.upper()

  return df_ubic


# Aplicación de la función
df_ubic: pd.DataFrame = normalizar_ubicaciones(df_horas)

# Mostrar las 10 primeras ubicaciones
df_ubic["ubicacion"].head(10)
````

````python
#@title **• Limpieza y normalización de patentes**

# Ver algunas patentes para entender los formatos
print("Patentes actuales:")
print(df_ubic["patente"].unique()[:20])
print(f"\nNulos actuales: {df_ubic['patente'].isnull().sum()}")
````

>Se ven patentes de distintos países (zona fronteriza Bélgica/Países Bajos/Alemania), así que los formatos son variados pero válidos. Lo que hay que hacer es:
>
> * Quitar caracteres especiales (como el - en BM -351)
> * Pasar a mayúsculas
> * Los nan (50 nulos) dejarlos como pd.NA


````python
#Normalización de patentes

import pandas as pd

def normalizar_patentes(df: pd.DataFrame) -> pd.DataFrame:
    """
    Normaliza la columna 'patente' del dataset:
    - Quita caracteres especiales
    - Elimina espacios
    - Convierte todo a mayúsculas
    - Reemplaza valores inválidos por pd.NA
    """

    df_pat: pd.DataFrame = df.copy()

    # Quitar caracteres especiales (deja letras, números y espacios)
    df_pat["patente"] = df_pat["patente"].str.replace(
        r"[^a-zA-Z0-9\s]", "", regex=True
    )

    # Pasar a mayúsculas
    df_pat["patente"] = df_pat["patente"].str.upper()

    # Eliminar espacios internos
    df_pat["patente"] = df_pat["patente"].str.replace(" ", "")

    # Reemplazar valores vacíos por pd.NA
    df_pat["patente"] = df_pat["patente"].replace(
        r"^\s*$", pd.NA, regex=True
    )

    return df_pat

    """
    Aplicando la función sobre el df ya con fechas y horas normalizadas
    """
df_pat: pd.DataFrame = normalizar_patentes(df_ubic)

print("Últimas 10 patentes normalizadas:")
df_pat[["patente"]].tail(10)
````

>Se realizó un proceso de limpieza sobre la columna patente con el objetivo de estandarizar su formato y eliminar inconsistencias. Las transformaciones aplicadas fueron:

>Eliminación de caracteres especiales: se conservaron únicamente letras, números y espacios mediante expresiones regulares.
Conversión a mayúsculas: para unificar el formato y evitar duplicados por diferencias de escritura.
Tratamiento de valores inválidos: los registros vacíos o compuestos solo por espacios se reemplazaron por pd.NA.

````python
#@title **• Eliminar filas con valores relevantes vacíos**

# Ver nulos por columna
print("Nulos por columna:\n")
print(df_pat.isnull().sum())
print(f"\nTotal de filas: {len(df_pat)}")
````

> Hay tres columnas con nulos:
>

> * **`velocidad_registrada`** (1957 nulos) → relevante, sin esto no podés calcular la infracción
> * **`radar_id`** (1002 nulos) → este es más discutible, es el identificador del radar pero no afecta el cálculo de la multa
>
>Las columnas relevantes para las multas son `patente` y `velocidad_registrada`. Sin esos datos la multa no tiene sentido. `radar_id` es información complementaria pero no esencial.


````python
columnas_relevantes: list[str] = [
  "patente",
  "velocidad_registrada"
]

# Cantidad de filas antes
filas_antes: int = len(df_pat)

# Eliminar filas con nulos en columnas relevantes
df_clean: pd.DataFrame = df_pat.dropna(subset=columnas_relevantes)

# Cantidad de filas eliminadas
filas_eliminadas: int = filas_antes - len(df_clean)

# Mostrar resultado con la leyenda que pide la consigna
print(f"Se eliminaron {filas_eliminadas} filas.")
print("Las columnas observadas son:")
for col in columnas_relevantes:
  print(f" - {col}")
````

````python
#@title **• Detectar y eliminar outliers**

# Ver estadísticas de las columnas numéricas
print("Estadísticas de las columnas numéricas:")
print(df_clean[["velocidad_registrada", "velocidad_maxima"]].describe())
````


> Los datos se ven bastante razonables pero vamos a usar el método estandar, IQR (rango intercuartílico) para detectar outliers, donde considera outlier todo valor que esté por debajo de `Q1 - 1.5*IQR` o por encima de `Q3 + 1.5*IQR.`


````python
columnas_numericas: list[str] = [
  "velocidad_registrada",
  "velocidad_maxima"
]

filas_antes: int = len(df_clean)

# Máscara para identificar outliers con método IQR
mascara_outliers: pd.Series = pd.Series(
  False, index=df_clean.index
)

for col in columnas_numericas:
  q1: float = df_clean[col].quantile(0.25)
  q3: float = df_clean[col].quantile(0.75)
  iqr: float = q3 - q1
  limite_inferior: float = q1 - 1.5 * iqr
  limite_superior: float = q3 + 1.5 * iqr

  mascara_outliers = mascara_outliers | (
    (df_clean[col] < limite_inferior) |
    (df_clean[col] > limite_superior)
  )

# Eliminar outliers
df_sin_outliers: pd.DataFrame = df_clean[~mascara_outliers]

filas_eliminadas: int = filas_antes - len(df_sin_outliers)

print(f"Se eliminaron {filas_eliminadas} filas.")
print("Las columnas observadas son:")
for col in columnas_numericas:
  print(f" - {col}")
````

> Dio 0 outliers, lo que tiene sentido, las velocidades van de 30 a 160 y los límites de 40 a 80, no hay valores extremos. El método IQR con esos rangos no detecta nada fuera de lo normal.


````python
#@title **• Crear columna exceso_velocidad_real**

df_sin_outliers["exceso_velocidad_real"] = (
  df_sin_outliers["velocidad_registrada"]
  - df_sin_outliers["velocidad_maxima"]
)

# Mostrar las 10 primeras patentes y su exceso
print("Top 10 de patentes y su exceso de velocidad:")
df_sin_outliers[["patente", "exceso_velocidad_real"]].head(10)
````

> Se ve que hay algunos con exceso negativo (como TL03 GOG con -2.0), lo que significa que iba más lento que el límite. Esos se van a eliminar más adelante.


````python
#@title **• Crear columna exceso_velocidad (con tolerancia del 5%)**

df_sin_outliers["exceso_velocidad"] = (
  df_sin_outliers["velocidad_registrada"]
  - df_sin_outliers["velocidad_maxima"]
  - (df_sin_outliers["velocidad_maxima"] * 0.05)
)

# Mostrar las 10 primeras patentes y su exceso
print("Top 10 de patentes y su exceso de velocidad con tolerancia 5%:")
df_sin_outliers[["patente", "exceso_velocidad"]].head(10)
````

> Se ve la diferencia con la tolerancia del 5%, el exceso baja un poco. Por ejemplo `WU 341 A` pasó de 47 a 44, y `TL03 GOG` pasó de -2 a -4 (todavía más lejos de ser infracción).


````python
#@title **• Eliminar filas sin infracción según exceso_velocidad**

filas_antes: int = len(df_sin_outliers)

# Conservar solo filas con exceso de velocidad positivo
df_infractores: pd.DataFrame = df_sin_outliers[
  df_sin_outliers["exceso_velocidad"] > 0
].copy()

filas_eliminadas: int = filas_antes - len(df_infractores)

print(f"Se eliminaron {filas_eliminadas} filas.")
````

## Ejercicio 04

> Puntos: 2

Definir una clase FineAnalyzer que cumpla con las siguientes funcionalidades:

- Reciba el DataFrame limpio como argumento de inicialización.

- Encapsule correctamente los datos.

- Implemente métodos para:

  - Ranking de las patentes más multadas (top 5). Debe retornar un DataFrame ordenado de mayor a menor y el índice debe comenzar desde 1, con las columnas de patentes y cantidad..

  - Ranking de los horarios donde se producen más multas (top 5). Debe retornar un DataFrame ordenado de mayor a menor y el índice debe comenzar desde 1, con las columnas de hora y cantidad.


  - Exceso promedio de velocidad. Debe retorna un flotante.

  - Exceso real promedio de velocidad . Debe retorna un flotante.

  - Contabilizar la cantidad de multas por cada  ubicación. Debe retornar un DataFrame ordenado por ubicación mostrando la ubicación y la cantidad.

- Crear el objecto e invocar cada uno de los métodos en celdas separadas.
  

````python
#@title **• Clase FineAnalyzer**


class FineAnalyzer:
    """
    Clase para analizar infracciones de velocidad.
    """

    def __init__(self, df: pd.DataFrame):
        """
        Inicializa la clase con el DataFrame limpio y encapsula los datos.
        """
        self.__data = df.copy()

    def ranking_patentes(self) -> pd.DataFrame:
        """
        Retorna el top 5 de patentes más multadas.
        """
        ranking = (
            self.__data["patente"]
            .value_counts()
            .head(5)
            .reset_index()
        )
        ranking.columns = ["patente", "cantidad"]
        ranking.index = ranking.index + 1
        return ranking

    def ranking_horarios(self) -> pd.DataFrame:
        """
        Retorna el top 5 de horarios con más multas.
        """
        ranking = (
            self.__data["hora"]
            .value_counts()
            .head(5)
            .reset_index()
        )
        ranking.columns = ["hora", "cantidad"]
        ranking.index = ranking.index + 1
        return ranking

    def exceso_promedio(self) -> float:
        """
        Retorna el exceso promedio de velocidad (con tolerancia).
        """
        return float(self.__data["exceso_velocidad"].mean())

    def exceso_real_promedio(self) -> float:
        """
        Retorna el exceso real promedio de velocidad.
        """
        return float(self.__data["exceso_velocidad_real"].mean())

    def multas_por_ubicacion(self) -> pd.DataFrame:
        """
        Contabiliza multas por ubicación y ordena alfabéticamente.
        """
        conteo = (
            self.__data["ubicacion"]
            .value_counts()
            .reset_index()
        )
        conteo.columns = ["ubicacion", "cantidad"]
        return conteo.sort_values(by="ubicacion")

"""
Inicialización del objeto
"""
analyzer = FineAnalyzer(df_infractores)
````

````python
#@title **• Exceso de velocidad agrupado por hora del día en formato HH:00**
# Copia de trabajo
df_hora: pd.DataFrame = df_infractores.copy()

# Convertir hora a datetime
df_hora["hora"] = pd.to_datetime(df_hora["hora"], format="%H:%M", errors="coerce")

# Agrupar por hora (redondeando a hora completa)
df_hora["hora_redondeada"] = df_hora["hora"].dt.floor("h")

# Calcular promedio de exceso de velocidad por hora
exceso_por_hora: pd.Series = (
    df_hora
    .groupby("hora_redondeada")["exceso_velocidad"]
    .mean()
    .sort_index()
)

# Crear gráfico de línea
plt.figure(figsize=(12,6))

plt.plot(
    exceso_por_hora.index,
    exceso_por_hora.values,
    color="red",
    marker="o"
)

for x, y in zip(exceso_por_hora.index, exceso_por_hora.values):
    plt.text(x, y, f"{y:.1f}", ha='center', va='bottom')

# Títulos
plt.title("Exceso de Velocidad Promedio por Hora", fontweight="bold")
plt.xlabel("Hora del día", fontweight="bold")
plt.ylabel("Exceso de velocidad promedio", fontweight="bold")

# Formatear eje X
plt.xticks(
    exceso_por_hora.index,
    [h.strftime("%H:%M") for h in exceso_por_hora.index],
    rotation=45
)

# Guardar gráfico
plt.savefig("urban_flow/data/interim/plots/hour.jpg", bbox_inches="tight")

# Mostrar gráfico
plt.show()
````
